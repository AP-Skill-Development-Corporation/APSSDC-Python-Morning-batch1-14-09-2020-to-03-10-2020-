```python
# inheritance:
- we can create  a class which uses all the properties and behaviour of another class
- new class is called as child,derived class,sub class
- and the one whose properties are acquired is known as  base class,parent class,super class
- re usability  of the code

A class--->parent
B class--->child

#syn for parent:
class parent_class():
    
#syn for child:
class childclass(parentclass):
    
#types of inheritances:
1)single inheritance
2)multiple
3)multi level
4)hierarchical
5)hybrid
```

```python
# single inheritance(one parent,one child)
class parent():
    a,b=10,45
    def display():
        print("iam from parent class")
        
class child(parent):
    c,d=23,67
    def show():
        print("iam from child")
```

```python
obj=parent
print(obj.a)
print(obj.b)
print(obj.display())
print(obj.c)
print(obj.d)
print(obj.show())
```

```python
obj=child
print(obj.a)
print(obj.b)
print(obj.display())
print(obj.c)
print(obj.d)
print(obj.show())
```

```python
# multiple inheritance(two parent,one child)
class father():
    def display1():
        print("iam from father class")

class mother():
    def display2():
        print("iam from mother class")
        
class child(father,mother):
    def show():
        print("iam from child")
```

```python
obj1=child
obj1.display1()
obj1.display2()
obj1.show()
```
